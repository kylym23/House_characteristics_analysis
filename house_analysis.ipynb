##### 10.1 Выберите дом клиенту
Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?



```python
df[(df['waterfront']==1) & (df['bathrooms']>=3) & (df['sqft_basement']>0)].shape[0]
```

##### 10.2 
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
condition=((df['waterfront']==1) | (df['view']==4)) & (df['condition']==5) & (df['yr_built']>=1980)
client_Choice=df[condition]
client_Choice

print('Диапазон от',client_Choice['price'].min(),'До', client_Choice['price'].max())
      
      
```

##### Какие характеристики недвижимости влияют на стоимость недвижимости?

```python
import numpy as np
```

```python
corr_matrix=df.corr()
corr_matrix=np.round(corr_matrix,1)
corr_matrix[np.abs(corr_matrix)<0.3]=0
corr_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```
